```python
import pandas as pd, pyarrow as pa, numpy as np, joblib
import os, time
import yfinance as yf
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import accuracy_score, roc_auc_score
from xgboost import XGBClassifier
import ta, plotly.graph_objects as go
```

### Load/Cache price data (csv + pickel)

```python
TICKERS = ["AAPL","MSFT","NVDA","AMZN","TSLA","META","JPM","XOM","NFLX","GOOGL"]
START, END = "2015-01-01", "2025-01-01"
DATA_DIR = "data_csv"       # folder with one CSV per ticker
PICKLE_PATH = "prices_2015_2025.pkl"
os.makedirs(DATA_DIR, exist_ok=True)

RUN_DOWNLOAD = False   # flip to True once to fetch, then back to False

if RUN_DOWNLOAD:
    for t in TICKERS:
        out = os.path.join(DATA_DIR, f"{t}_{START}_{END}.csv")
        if os.path.exists(out):
            print(f"[skip] {t} (already cached)")
            continue
        print(f"Downloading {t} ...")
        df = yf.download(t, start=START, end=END, auto_adjust=True, threads=False, progress=False)
        if df is None or df.empty:
            print(f"  -> empty / rate-limited for {t}, try later")
        else:
            df = df.rename(columns=str.title).reset_index()
            df["Symbol"] = t
            df.to_csv(out, index=False)
            print(f"  -> saved {out}")
        time.sleep(5)  # gentle pause
else:
    print("Download skipped (RUN_DOWNLOAD=False). Using cached files if present.")
```

### Assemble panel & Sanity check

```python
panel = None
# Prefer fast Pickle if it exists
if os.path.exists(PICKLE_PATH):
    panel = pd.read_pickle(PICKLE_PATH)
    print(f"Loaded Pickle -> {PICKLE_PATH}")
else:
    # Build from CSVs
    files = [os.path.join(DATA_DIR, f) for f in os.listdir(DATA_DIR) if f.endswith(".csv")]
    if not files:
        print("No cached CSVs found. Set RUN_DOWNLOAD=True in the previous cell once to fetch.")
    else:
        dfs = [pd.read_csv(p, parse_dates=["Date"]) for p in files]
        panel = pd.concat(dfs, ignore_index=True).sort_values(["Symbol","Date"])
        panel.to_pickle(PICKLE_PATH)
        print(f"Loaded {len(files)} CSVs and saved Pickle -> {PICKLE_PATH}")

panel = panel.copy()
panel["Date"] = pd.to_datetime(panel["Date"], errors="coerce")

num_cols = ["Open","High","Low","Close","Adj Close","Volume"]
for c in num_cols:
    if c in panel.columns:
        # If any CSV had commas or spaces, strip them then coerce:
        if panel[c].dtype == "object":
            panel[c] = (
                panel[c]
                .astype(str)
                .str.replace(",", "", regex=False)
                .str.strip()
            )
        panel[c] = pd.to_numeric(panel[c], errors="coerce")

# Quick sanity check
if isinstance(panel, pd.DataFrame):
    print(panel.head(3))
    print("Rows:", len(panel),
          "| Symbols:", panel['Symbol'].nunique(),
          "| Dates:", panel['Date'].min(), "→", panel['Date'].max())
```

```python
required_cols = ["Date","Symbol","Open","High","Low","Close","Volume"]
missing = [c for c in required_cols if c not in panel.columns]
if missing:
    raise ValueError(f"panel is missing columns: {missing}")

data = (
    panel[required_cols]
    .copy()
    .sort_values(["Symbol","Date"])
    .reset_index(drop=True)
)

# Ticker list (from your data)
tickers = data["Symbol"].unique().tolist()
print("Tickers:", tickers[:10], "… (total:", len(tickers), ")")
print(data.head())
```

### Feature engineering

```python
import ta
def add_features(g: pd.DataFrame, N: int = 5, tau: float = 0.0) -> pd.DataFrame:
    g = g.sort_values("Date").copy()
    
    # Price-base momentum
    g["ret1"]   = g["Close"].pct_change()
    g["ret5"]   = g["Close"].pct_change(5)
    g["ret10"]   = g["Close"].pct_change(10)

    # EMAs & slope
    g["ema10"]  = ta.trend.EMAIndicator(g["Close"], 10).ema_indicator()
    g["ema20"]  = ta.trend.EMAIndicator(g["Close"], 20).ema_indicator()
    g["ema50"]  = ta.trend.EMAIndicator(g["Close"], 50).ema_indicator()
    g["ema50_slope"] = g["ema50"].pct_change(5)

    # Oscillators
    rsi = ta.momentum.RSIIndicator(g["Close"], 14)
    g["rsi14"] = rsi.rsi()
    stoch = ta.momentum.StochasticOscillator(g["High"], g["Low"], g["Close"], 14, 3)
    g["stoch_k"] = stoch.stoch()
    g["stoch_d"] = stoch.stoch_signal()
    
    # MACD
    macd        = ta.trend.MACD(g["Close"])
    g["macd"]   = macd.macd()
    g["macds"]  = macd.macd_signal()
    g["macd_hist"]= macd.macd_diff()

    
    # Volatility
    atr = ta.volatility.AverageTrueRange(g["High"], g["Low"], g["Close"], 14)
    g["atr"]     = atr.average_true_range()
    g["atr_pct"] = g["atr"] / g["Close"]
    bb = ta.volatility.BollingerBands(g["Close"])
    g["bb_w"]    = bb.bollinger_wband()
    
    # Volume features
    if "Volume" in g:
        g["vol_chg"] = g["Volume"].pct_change().clip(-5, 5)
        g["rv20"]    = g["Volume"] / (g["Volume"].rolling(20).mean() + 1e-9)

    # Label: N-bar forward return, optionally thresholded
    g["fwd_ret"] = g["Close"].shift(-N) / g["Close"] - 1
    g["y"] = (g["fwd_ret"] > tau).astype(int)

    return g

data = (
    panel.sort_values(["Symbol","Date"])
         .groupby("Symbol", group_keys=False)
         .apply(lambda g: add_features(g).assign(Symbol=g.name))
         .dropna()
         .reset_index(drop=True)
)

data = data.dropna().reset_index(drop=True)
print("Rows after features:", len(data))
```

### Train/Val/Test

```python
syms = data["Symbol"].unique()
train_syms = syms[: int(0.8*len(syms))]
test_syms  = syms[int(0.8*len(syms)):]

train = data[(data.Symbol.isin(train_syms)) & (data.Date <  "2020-01-01")]
valid = data[(data.Symbol.isin(train_syms)) & (data.Date >= "2020-01-01") & (data.Date < "2023-01-01")]
test  = data[(data.Symbol.isin(test_syms))  & (data.Date >= "2023-01-01")]
features = ["ret1","ema10","ema20","ema50","macd","macds","atr","bb_w"]
```

### Per symbol Normalization

```python
def fit_stats(df, cols):
    return {s:(g[cols].mean(), g[cols].std().replace(0,1))
            for s,g in df.groupby("Symbol")}
def apply_stats(df, cols, stats):
    out = df.copy()
    for s,g in out.groupby("Symbol"):
        mu, sd = stats.get(s, (g[cols].mean(), g[cols].std().replace(0,1)))
        out.loc[g.index, cols] = (g[cols]-mu)/sd
    return out

stats = fit_stats(train, features)
train_n = apply_stats(train, features, stats)
valid_n = apply_stats(valid, features, stats)
test_n  = apply_stats(test,  features, stats)

X_tr, y_tr = train_n[features].values, train_n["y"].values
X_va, y_va = valid_n[features].values, valid_n["y"].values
X_te, y_te = test_n[features].values,  test_n["y"].values
```

## Best params for modell

```python
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV

xgb_clf = XGBClassifier(
    objective="binary:logistic",
    eval_metric="logloss",
    n_estimators=500,
    use_label_encoder=False
)

param_dist = {
    "max_depth": [3, 4, 5, 6],
    "learning_rate": [0.01, 0.03, 0.05, 0.1],
    "subsample": [0.6, 0.7, 0.8, 0.9],
    "colsample_bytree": [0.6, 0.7, 0.8, 0.9],
    "min_child_weight": [1, 3, 5],
    "gamma": [0, 0.5, 1],
}

search = RandomizedSearchCV(
    xgb_clf,
    param_distributions=param_dist,
    n_iter=20,
    scoring="roc_auc",
    cv=3,
    verbose=1,
    random_state=42,
    n_jobs=1         # <- run in main process, no worker imports
)

search.fit(X_tr, y_tr)
print("Bästa AUC:", search.best_score_)
print("Bästa parametrar:", search.best_params_)
```

### Train XGBoost (CV on Train -> best_rounds best_rounds -> fit)

```python
import xgboost as xgb

# --- DMatrices ---
Xtr = np.asarray(X_tr, dtype=np.float32);  ytr = np.asarray(y_tr)
Xva = np.asarray(X_va, dtype=np.float32);  yva = np.asarray(y_va)

dtrain = xgb.DMatrix(Xtr, label=ytr)
dvalid = xgb.DMatrix(Xva, label=yva)

params = {
    "objective": "binary:logistic",
    "eval_metric": "logloss",
    "max_depth": 6,
    "eta": 0.05,                 # learning_rate -> eta for xgb.train
    "subsample": 0.9,
    "colsample_bytree": 0.7,
    "min_child_weight": 1,
    "gamma": 1,
    "lambda": 1.0,
    "alpha": 0.0,
    "seed": 42,
}

# --- CV on TRAIN ONLY to get best_rounds ---
cv = xgb.cv(
    params=params,
    dtrain=dtrain,
    num_boost_round=2000,
    nfold=5,
    early_stopping_rounds=50,
    verbose_eval=False,
    stratified=True,
    seed=42,
)
best_rounds = len(cv)
print("best_rounds from CV:", best_rounds)

# --- Train on TRAIN ONLY with best_rounds, evaluate on VAL ---
bst = xgb.train(params, dtrain, num_boost_round=best_rounds)

p_val = bst.predict(dvalid)
val_auc = roc_auc_score(y_va, p_val)
val_acc = accuracy_score(y_va, (p_val >= 0.5).astype(int))
print("Val AUC:", round(val_auc, 3))
print("Val Acc:", round(val_acc, 3))
```

### Threshold Band Search (Validation) - Long/short

```python
import numpy as np
import pandas as pd

def eval_long_short(df, probs, thr_hi=0.60, thr_lo=0.40, fee_bps=1.0, use_flat_band=True):
    """
    Long/short evaluator with 1-bar delay and transaction costs.
    If use_flat_band=True, positions are:
        +1 if p>=thr_hi, -1 if p<=thr_lo, else 0 (flat)
    If use_flat_band=False, positions are:
        +1 if p>=thr_hi, else -1 (always in market)
    Returns dict with cumret, sharpe, maxdd, daily series.
    """
    d = df[["Date","Symbol","Close"]].copy().reset_index(drop=True)
    d["p"] = np.asarray(probs)
    d["ret"] = d.groupby("Symbol")["Close"].pct_change().fillna(0.0)

    if use_flat_band:
        pos = np.where(d["p"] >= thr_hi, 1, np.where(d["p"] <= thr_lo, -1, 0))
    else:
        pos = np.where(d["p"] >= thr_hi, 1, -1)

    # 1-bar delay to act on signals
    pos_lag = pd.Series(pos).shift(1).fillna(0).values

    # transaction costs on changes (magnitude of change)
    flips = np.abs(pd.Series(pos_lag).diff().fillna(0).values)
    cost = flips * (fee_bps / 10000.0)

    strat = pos_lag * d["ret"].values - cost
    daily = pd.Series(strat, index=d["Date"]).groupby(level=0).mean().sort_index()

    # metrics
    eq = (1.0 + daily).cumprod()
    ann = 252
    mu, sd = daily.mean(), daily.std() + 1e-12
    sharpe = np.sqrt(ann) * mu / sd
    maxdd = (eq / eq.cummax() - 1.0).min()
    cumret = (eq.iloc[-1] - 1.0) if len(eq) else 0.0

    # simple buy&hold (equal-weight across symbols per day)
    bh = d.groupby("Date")["Close"].mean()
    bh = (bh / bh.iloc[0]).iloc[-1] - 1.0

    return {
        "thr_hi": float(thr_hi),
        "thr_lo": float(thr_lo),
        "cumret": float(cumret),
        "sharpe": float(sharpe),
        "maxdd": float(maxdd),
        "bh": float(bh),
        "daily": daily,
    }

def pick_best_threshold_longshort(valid_df, probs, fee_bps=1.0,
                                  grid_hi=None, grid_lo=None,
                                  metric="sharpe", use_flat_band=True):
    """
    Grid-search thr_lo/thr_hi on VALID for long/short. Picks best by metric.
    """
    if grid_lo is None:
        grid_lo = np.arange(0.30, 0.50, 0.01)  # low threshold candidates
    if grid_hi is None:
        grid_hi = np.arange(0.50, 0.71, 0.01)  # high threshold candidates

    best = None
    for lo in grid_lo:
        for hi in grid_hi:
            if lo >= hi:
                continue
            m = eval_long_short(valid_df, probs, thr_hi=hi, thr_lo=lo,
                                fee_bps=fee_bps, use_flat_band=use_flat_band)
            score = m["sharpe"] if metric == "sharpe" else m["cumret"]
            if (best is None) or (score > best["score"]):
                m["score"] = float(score)
                best = m
    return best

# --- RUN on your VALID set ---
# valid_n: your validation DF aligned with p_val
best_ls = pick_best_threshold_longshort(
    valid_n, p_val,
    fee_bps=1.0,
    metric="sharpe",          # or "return" to maximize total return
    use_flat_band=True        # set False to be always-in-market long/short
)

print(f"Best long/short thresholds on VALID:"
      f" thr_lo={best_ls['thr_lo']:.3f}, thr_hi={best_ls['thr_hi']:.3f}")
print(f"  Sharpe: {best_ls['sharpe']:.2f} | CumRet: {best_ls['cumret']*100:.2f}%"
      f" | MaxDD: {best_ls['maxdd']*100:.2f}% | Buy&Hold: {best_ls['bh']*100:.2f}%")

# --- COMMENTED OUT TEST EVALUATION ---
# p_test = bst.predict(xgb.DMatrix(np.asarray(X_te, np.float32)))
# test_ls = eval_long_short(test_n, p_test,
#                           thr_hi=best_ls['thr_hi'], thr_lo=best_ls['thr_lo'],
#                           fee_bps=1.0, use_flat_band=True)
# print(f"TEST long/short:"
#       f" Sharpe: {test_ls['sharpe']:.2f} | CumRet: {test_ls['cumret']*100:.2f}%"
#       f" | MaxDD: {test_ls['maxdd']*100:.2f}% | Buy&Hold: {test_ls['bh']*100:.2f}%")
```

### Visualize Signals

```python
import matplotlib.pyplot as plt

# --- CONFIG: set these before running ---
SYMBOL        = None          # e.g. "AAPL" or "XOM"; if None, auto-picks most common in VALID
START         = "2020-01-01"   # e.g. "2023-01-01" or None to ignore
END           = None           # e.g. "2024-12-31" or None to ignore
LAST_N        = None           # e.g. 250 to show last ~1y; if set, overrides START/END
THR_HI        = 0.57           # go LONG when p >= THR_HI
THR_LO        = 0.33           # go SHORT when p <= THR_LO
USE_FLAT_BAND = True           # True: flat between LO/HI; False: always-in-market long/short
FEE_BPS       = 1.0            # transaction cost per position change (basis points)

# 1) Build working frame (align probs, positions)
_plot_df = valid_n.copy().reset_index(drop=True)
_plot_df["Date"] = pd.to_datetime(_plot_df["Date"], errors="coerce")
_plot_df["p"]    = p_val

# Select symbol (auto if not provided)
if SYMBOL is None:
    counts = _plot_df["Symbol"].value_counts()
    if counts.empty:
        raise ValueError("No symbols found in validation data.")
    SYMBOL = counts.index[0]

view = _plot_df[_plot_df["Symbol"] == SYMBOL].dropna(subset=["Date","Close"]).sort_values("Date").copy()

# 2) Positions from thresholds
if USE_FLAT_BAND:
    view["pos"] = np.where(view["p"] >= THR_HI, 1, np.where(view["p"] <= THR_LO, -1, 0))
else:
    view["pos"] = np.where(view["p"] >= THR_HI, 1, -1)

# 3) Apply date window
if LAST_N is not None:
    view = view.tail(int(LAST_N))
else:
    if START is not None:
        view = view[view["Date"] >= pd.to_datetime(START)]
    if END is not None:
        view = view[view["Date"] <= pd.to_datetime(END)]

if view.empty:
    raise ValueError(f"No data to plot for {SYMBOL} in the selected window. "
                     f"Try another SYMBOL or adjust START/END/LAST_N.")

# 4) Entry/exit markers (where position changes)
view["flip"]       = view["pos"].diff().fillna(view["pos"])
view["long_entry"] = view["flip"].eq(1)   # moved to +1
view["short_entry"]= view["flip"].eq(-1)  # moved to -1
view["flat_entry"] = view["flip"].eq(0) & view["pos"].eq(0)  # (only meaningful if using flat band)

# 5) Plot price + entries
plt.figure(figsize=(12, 6))
plt.plot(view["Date"], view["Close"], label=f"{SYMBOL} Close")
# Long entries
plt.scatter(view.loc[view["long_entry"], "Date"],
            view.loc[view["long_entry"], "Close"], marker="^", label="Long entry")
# Short entries
plt.scatter(view.loc[view["short_entry"], "Date"],
            view.loc[view["short_entry"], "Close"], marker="v", label="Short entry")
# Optional: show when we go flat
if USE_FLAT_BAND and view["flat_entry"].any():
    plt.scatter(view.loc[view["flat_entry"], "Date"],
                view.loc[view["flat_entry"], "Close"], marker="o", label="Flat")

plt.title(f"{SYMBOL} — Long/Short Signals (hi={THR_HI}, lo={THR_LO}, flat_band={USE_FLAT_BAND})")
plt.xlabel("Date"); plt.ylabel("Price")
plt.legend(); plt.grid(True, linewidth=0.3)
plt.show()

# 6) Quick strategy stats in the plotted window (long/short, 1-bar delay, with costs)
view["ret"] = view["Close"].pct_change().fillna(0.0)
pos_lag     = view["pos"].shift(1).fillna(0.0)
flips       = pos_lag.diff().abs().fillna(pos_lag.abs())  # charge when position changes at start or later
cost        = flips * (FEE_BPS / 10000.0)
strat       = pos_lag * view["ret"] - cost

eq = (1 + strat).cumprod()
cumret = (eq.iloc[-1] - 1) * 100 if len(eq) else 0.0
n_long_entries  = int(view["long_entry"].sum())
n_short_entries = int(view["short_entry"].sum())

print(f"Bars: {len(view)} | Long entries: {n_long_entries} | Short entries: {n_short_entries}")
print(f"Cumulative strategy return in window: {cumret:.2f}% (long/short, 1-bar delay, fee {FEE_BPS} bps)")
```
